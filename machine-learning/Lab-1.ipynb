### Какое самое популярное женское имя на корабле?

Для того, чтобы опередлить самое популярное женское имя, нужно из каждой записи в колонке 'Name', соответствующей женщине, выбрать все ее имена. То есть нужно исключить фамилию и имя мужа, девичью фамилию, и оставить только первое, второе и т.д. личные имена.

```python
import re
names = titanik_df[titanik_df['Sex'] == 'female']['Name'].tolist()
name_list = []
for name in names:         
    name = name.replace('"', '');
    reg = re.match("[a-z A-Z'\-]+, Mrs.[a-z A-Z']+[(]([a-z A-Z.]+)[)]",name)
    if reg:
        cur_names = reg.group(1).split()
        if len(cur_names) > 1:
            cur_names = cur_names[:-1]
    else:
        reg = re.match("[a-z A-Z'\-]+, (?:Mrs.|Miss.|Ms.|Mme.|Mlle.)([a-z A-Z]+)",name)
        if reg:
            cur_names = reg.group(1).split() 
    name_list += cur_names
```

Теперь в списке name_list хранятся все вожможные женские имена, которые только встречаются на борту Титаника. Подсчитаем количество вхождений каждого имени и построим диаграмму для наиболее популярных имен (которые носят более пяти человек).

```python
names_df = pd.DataFrame({'number' : pd.Series(name_list).value_counts(ascending=True)})
names_df[names_df['number'] > 5].plot(kind = 'barh', legend = False, figsize = (6, 6),fontsize=15, colormap = 'Set1')
xlabel('Passengers number')
```

Из диграммы видно, что самым популярным женским имененм на корабле было имя Mary.

### Какие титулы имели пассажиры (примеры титула — Mrs., Mr.), едущие различными классами?

Сначала добавим набор данных в titanik_dataset еще один признак - титул 'Title', значения которого получим из признака 'Name'

```python
titles_list = []
def GetTitle(name):
    reg = re.match("[a-z A-Z'\-]+, ([A-Z][a-zA-Z]+).", name) 
    if reg:
        return reg.group(1) + '.'
    return NaN
titanik_df['Title'] = titanik_df['Name'].map(GetTitle)
titanik_df[0:6]
```

После этого построим круговые диаграммы, откажающие титулы пассажиров, едущих различными классами.

```python
fig = figure(figsize = (60, 20))
rcParams['font.size'] = 50.0
fig.fontsize = 20
titles = titanik_df['Title'].value_counts().keys()
colors = plt.cm.Set1(np.linspace(0, 1, len(titles.tolist())))
colors_dict = dict(zip(titles, colors))

def GetColors(cur_titles):
    cur_colors = []
    for title in cur_titles:
        cur_colors.append(colors_dict[title])
    return cur_colors

for cl in range(1, 4):
    f = subplot(1, 3, cl)
    title(str(cl) + ' class')
    data = titanik_df[titanik_df['Pclass'] == cl]['Title'].value_counts()
    cur_titles = data.keys()
    axis('equal')
    if cl == 1:
        pie_ = pie(data.tolist(), 
                   colors = GetColors(cur_titles), 
                   labeldistance=1.1,
                   autopct=lambda(p): '{:.0f}%'.format(p) if p >= 1 else ''
                   );
        legend(loc=(-0.5,0), shadow=True, labels = cur_titles)
    else:
        pie_ = pie(data.tolist(), 
                   colors = GetColors(cur_titles), 
                   labels = cur_titles, 
                   labeldistance=1.1,
                   autopct=lambda(p): '{:.0f}%'.format(p) if p >= 1 else ''
                   );
```

### Есть ли зависимость между классом и номером билета?

Номера билетов пассажиров Титаника имеют различную структуру, но все нормера имеют числовую часть. Для того чтобы привести их к единому виду выделим это число. Будем рассматривать зависимоть между числом, полученным из номера билета, и классом пассажира.

```python
def GetTicketNumber(ticket):
    strs = ticket.split()
    if len(strs) == 2:
        number = (strs[1])
    else:
        number = (strs[0])
    return number
titanik_df['Ticket number'] = titanik_df['Ticket'].map(GetTicketNumber)
titanik_df[['Pclass', 'Ticket number', 'Ticket']][0:10]
```

На основе данных в таблице можно выдвинуть гипотезу, что большинство чисел в колонке Ticket number, начинаются с цифры, соответствующей номеру класса. Посчитаем число строк удовлетворяющих данной гипотезе.

```python
true_count = 0
false_count = 0
for index, p in titanik_df.iterrows():
    if p['Ticket number'].startswith(str(p['Pclass'])):
        true_count+=1
    else:
        false_count+=1
total = len(titanik_df)        
print 'Success', true_count, '/', total, '{:.1f}%'.format((float(true_count) / total)*100)
print 'Failure', false_count, '/', total, '{:.1f}%'.format((float(false_count) / total)*100) 
```

В итоге получилось, что 72 % всех номеров билетов начинаются с номера класса, следовательно существует зависимость между классом и номером билета.

### Верно ли, что чаще выживали пассажиры с более дорогими билетами?

Упорядочим всех пассажиров по цене билета, а затем разобьем их на группы по 27 человек и посчитаем число выживших пассажиров в каждой группе. Представим результаты в виде диаграммы, каждый столбец диаграммы показывает, сколько из 27 пассажиров каждой группы остались живы. Так как изначально все пассажиры были разделены на равные группы, то по высоте столбцов можно оценить зависимость выживаимости от цены билета.

```python
rcParams['font.size'] = 10.0
figure(figsize(15, 7))
data = titanik_df.sort('Fare', ascending = 1)
size = len(data)
data = data.reset_index()
data = data.reindex(index=range(0,size))
group_size = 27
survived = []
labels = []
row = 0
for number in xrange(0, size, group_size):
    first_row = row
    second_row = row + group_size
    dta = data[first_row : second_row]
    survived.append(len(dta[dta['Survived'] == 1]))
    row += group_size
    labels.append("({0:.1f}-{1:.1f})".format(data['Fare'][first_row], data['Fare'][second_row-1]))

data = pd.Series(survived, labels)
data.plot(kind = 'bar',
          colormap = 'Set1')
xlabel('Fare')
ylabel('Survived passengers number')

```

На построенной диаграмме видно, что с увеличением цены билета увеличивается и число выживших пассажиров. То есть пассажиры с более дорогими билетами, действительно, выживали чаще. Наверное, это можно объяснить тем, что они занимали места на верхней палубе, с которой было проще выбраться.

### Закономерности, точно описывающие группу выживших пассажиров

#### Закономерность 1

Если женщина, не имеет родных на корабле, и купила дорогой (Fare > 30) билет, то она выживет.

```python
data = titanik_df[(titanik_df['SibSp'] == 0) & (titanik_df['Parch'] == 0) & (titanik_df['Sex'] == 'female') & (titanik_df['Fare'] >= 30)]
print 'Not survived', len(data[data['Survived'] == 0])
print 'Passangers number', len(data[data['Survived'] == 1])
```

Эта закономерность кажется логической, так как, во-первых, как всем известно, женщин и детей спасают в первую очередь, во-вторых, пассажирам, не имеющих родных на корабле было легче спастись, ведь им не нужно было беспокоиться за любимых членов семьи, в-третьих, мы уже обнаружили закономерность, связаную с тем, что пассажиры с дорогими билетами выживают чаще.
Данная закономерность сложная, так как зависит от многих условий.

Если пассажир ехал в каюте, в которой нет погибиших пассажиров, то он выживет.

```python
index = []
data = []
good_cabins = []
figure(figsize(30, 10))
t = titanik_df.copy()
t['Survived'] = t['Survived'].map(lambda x:'Survived' if x == 1 else 'Not survived')
for ind, group in t.groupby('Cabin'):
    index.append(ind)
    passengers = pd.value_counts(group['Survived'])
    if not(0 in passengers.keys()):
        good_cabins.append(ind)
    data.append(passengers)
df = pd.DataFrame(data, index)
df.plot(kind='bar', stacked=True, colormap = 'Set1', legend = True)
plt.xlabel('Cabin')
plt.ylabel('Passangers number')
```
